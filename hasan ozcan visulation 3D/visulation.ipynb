```python
import json
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Path to your JSON file
json_file_path = 'C:\\Users\\hasan\\OneDrive\\Masast\\ieco\\file1\\dsm.json'

# Load JSON data
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Convert data to NumPy array
data_array = np.array(data)

# Print the data to verify
print("Data as NumPy array:\n", data_array)

# Separate columns if needed
x = data_array[:, 0]  # X coordinates
y = data_array[:, 1]  # Y coordinates
z = data_array[:, 2]  # Z values

# Example of some analysis with SciPy
mean_z = np.mean(z)
std_z = np.std(z)

print(f"Mean Z value: {mean_z}")
print(f"Standard Deviation of Z values: {std_z}")

# Example of linear regression using SciPy
slope, intercept, r_value, p_value, std_err = stats.linregress(x, z)

print(f"Linear Regression - Slope: {slope}, Intercept: {intercept}")

# Plotting the data
plt.scatter(x, z, label='Data points')
plt.plot(x, slope * x + intercept, color='red', label='Fit Line')
plt.xlabel('X coordinate')
plt.ylabel('Z value')
plt.title('Scatter Plot with Fit Line')
plt.legend()
plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Path to your JSON file
json_file_path = 'C:\\Users\\hasan\\OneDrive\\Masast\\ieco\\file1\\dsm.json'

# Load JSON data
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Convert data to NumPy array
data_array = np.array(data)

# Separate columns
x = data_array[:, 0]  # X coordinates
y = data_array[:, 1]  # Y coordinates
z = data_array[:, 2]  # Z values

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o')

# Labels and title
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Z Value')
ax.set_title('3D Scatter Plot of Data')

# Show plot
plt.show()
```
